```python
import sys

class VariableTracker:
    def __init__(self):
        self.previous_values = {}
    
    def trace(self, frame, event, arg):
        # Get the current values of all local variables in the frame
        current_values = frame.f_locals
        if event == "line" and __name__ == "__main__":
            print(frame.f_code.co_name)
        
        # Compare the current values with the previous values
        for variable_name, current_value in current_values.items():
            print(frame.f_locals)
            previous_value = self.previous_values.get(variable_name)
            #if previous_value != current_value:
            #    print(f"{variable_name} changed: {previous_value} -> {current_value}")
        
        # Update the previous values with the current values
        self.previous_values.update(current_values)
        
        return self.trace
    
    def start_tracking(self):
        sys.settrace(self.trace)
        
    def stop_tracking(self):
        sys.settrace(None)
```

```python


def main():
    x = 1
    y = 2
    z = x + y
    y = 3
    z = x + y

if __name__ == '__main__':
    tracker = VariableTracker()
    tracker.start_tracking()
    main()
    tracker.stop_tracking()
```
